```python
# Define and initialize variables
numPeople = 10  # Number of people that are having omelette.
numEggs = 0 # Number of eggs needed. We don't know the value yet, but initialize to 0
numPotatoes = 0  # Number of potatoes needed. We don't know the value yet, but initialize to 0
numOnions = 0  # Number of onions needed. We don't know the value yet, but initialize to 0

withSalt = True  # If we want to add salt, this variable should be true.
withOnion = True  # If we want to add onion, this variable should be true.
withLevadura = True # If we want to add levadura, this variable should be true.

# Calculate the number of potatoes and eggs needed.
numPotatoes = numPeople + 1  # One potato per person, + 1 extra.
numEggs = numPeople + 2  # One egg per person, + 2 extra.

# Tell the cook how many potatoes and eggs are needed.
print("Ingredients:")
print("To cook a 'Tortilla' for ", numPeople, " people, you need:")
print("   - ", numPotatoes, " potatoes")
print("   - ", numEggs, " eggs")

# If onion is needed, inform the cook.
if withOnion:
    numOnions = 1
    print("   - ", numOnions, " onion(s)")


# If salt is needed, inform the cook.
if withSalt:
    print("   - Salt")

# If levadura is needed, inform the cook.
if withLevadura:
    print("   - Levadura")

# Start cooking
print("Let's start cooking.")

# Step 1: Peel and chop the potatoes.
print("Peel and chop the potatoes:")

for i in range(numPotatoes):
    print("   - Peel and chop potato number ", i)


# Step 2: Heat the oil and fry the potatoes.
print("In a pan, pour enough oil to cover the potatoes and put it to heat.")
print("When the oil is hot, add the potatoes.")

# Step 3: Crack and beat the eggs.
print("While we cook the potatoes, crack and beat the eggs:")

for i in range(numEggs):
    print("   - Crack and pour into a bowl egg number ", i)


# If salt is needed, add it to the mixture.
if withSalt:
    print("Add salt to the eggs.")
    
    
# Beat the eggs.
print("Beat the eggs.")


# Step 4: When the potatoes are ready, mix the with the eggs.
print("When ready, take the potatoes off the pan, rinse them, and mix with the eggs.")

# If Levadura is needed, add it to the mixture.
if withLevadura:
    print("Add Levadura to the mix.")

# Step 5: If onion is needed, fry and add it to the mixture.
if withOnion:
    print("Chop the onion. Fry it till golden and add it to the mixture.")

# Step 5: Cook the mixture to obtain the tortilla.
print("Pour the potato and egg mixture into the pan.")
print("   - Cook for 5 minutes.")
print("   - Carefully flip the tortilla.")
print("   - Heat for 5 more minutes.")

print("Your tortilla is ready to eat!")
```
